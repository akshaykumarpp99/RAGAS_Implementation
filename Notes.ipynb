## Ragas

- It ia a Quantitative measure used to evaluate performance of AI Applications.
- It provides the numerical biases for comparision, optimization and decision making.


### These Metrics are crucial for:

1. Component Selection: 
    - Metrics can be used to compare different components of the AI Appication like LLM, Retriever and Agent Configuration etc with owna data and select the best one.
2. Error Diagnosis and Debugging: 
    - Metrics help us to identify which part of the application is causing errors or suboptimal performance, making it easier to refine.
3. Continuous Monitor and Maintenance: 
    - Mertics enable the tracking of AI Application performance over time, helphing to detect and respond issues such as data drift, model degradation, or changing user requirements.

## Different Types of metrics:



## LLM Based Metrics:

* These metrics uses LLM underneath to do the evaluation. 
* There might be one or more LLM calls to decide the final score.
* These metrics can be somewhat non deterministic as the LLM might not generate the same response for same input.
* One advantage is these metrics has more accurate and closer to human evaluation.

## Non LLM Based Metrics:

* These Metrics dont use LLM's underneath to do evaluation.
* These Metrics are deterministic and can be used to evaluate the performance of AI Application without using LLM.
* These metrics rely on traditional methods to evaluate the performance such as string similarity, BLEU Score.
* These metrics have lower coreleation with human evaluation.

### Metric is classified into 2 categories based on type of data they evaluate

* Single turn: Single turn interaction between user and the AI
* Multi turn: Multi turn interaction between user and AI.

## Metric Design Principles:

1. Single Aspect Focus
2. Intuitive and Interpretable
3. Effective Prompt Flows
4. Robustness: 
    - Ensure that LLM-based metric include few shot examples that reflects the desired output.
    - By Providing context and guidance to follow, which will increase the robustness of the metric.
5. Consistent Scoring Ranges: It is crucial to normalize the metric score values and to ensure they fall in a specific range i.e 0-1.
